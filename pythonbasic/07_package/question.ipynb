### *문제 2: **카피바라의 업무 관리 패키지***

    - `*task_tracking.py`: 작업 상태를 관리하는 모듈*
        

    - `*reporting.py`: 업무 보고를 관리하는 모듈*

    ```python
    작업 '코드 리뷰' 시작됨.
    작업 '코드 리뷰' 종료됨.
    '코드 리뷰' 작업 보고서 생성됨.
    ```

````python
from work_management import task_tracking as tt, reporting as rp


print(tt.start_task("코드 리뷰"))
print(tt.end_task("코드 리뷰"))
print(rp.generate_report("코드 리뷰"))
````

````python
from work_management.task_tracking import start_task as st, end_task as et
from work_management.reporting import generate_report as gr

print(st("코드 리뷰"))
print(et("코드 리뷰"))
print(gr("코드 리뷰"))
````
