**"카피바라의 주말 계획은?"**

1. 사용자에게 "오늘 날씨가 어떤가요? (맑음/비/눈): "를 입력받으세요.

```python
text = input("오늘 날씨가 어떤가요?(맑음/비/눈) : ")

match text:
    case '맑음':
        print("한강에 가서 피크닉을 즐기는 척하며 일을 한다.")
    case '비':
        print("집에서 빗소리를 ASMR 삼아 일을 한다.")
    case '눈':
        print("전기장판 안에서 귤을 까먹으며 일을 한다.")
    case _:
        print("암막 커튼 때문에 일어나지 못했다. 푹 잘 수 있군, 럭키비키잖아?")
```

3. 손님이 0이 될 때까지 **"남은 손님 수: [customers]명"**을 출력하세요.

1. 변수 `work_hours`에 일한 시간(정수)을 입력받으세요.
2. `while`문을 사용해 일한 시간만큼 반복하며 `pay`에 시급을 더하세요.
3. 최종 알바비를 **"오늘의 알바비는 [pay]원입니다."**로 출력하세요.

1. 주문 리스트 `orders`에 `'아메리카노', '카페라떼', '핫초코', '모르겠어요', '에스프레소'`를 저장하세요.

**"카피바라는 짝수 테이블만 청소하고 싶다."**

카피바라는 홀수 테이블은 건너뛰고 짝수 테이블만 청소하려고 합니다.

카피바라는 홀수 테이블은 건너뛰고 짝수 테이블만 청소하려고 합니다.

1. 테이블 리스트는 `1`부터 `6`까지의 숫자로 구성되어 있습니다.
2. `for`문을 사용해 각 테이블 번호를 확인하세요.
3. 테이블 번호가 **홀수**이면 해당 테이블을 건너뛰세요.
4. 짝수 테이블인 경우 `"[n번 테이블] 청소 완료!"`를 출력하세요.

```python
table_list = range(1,7)
for table_num in table_list:
    if table_num % 2 != 0:
        continue
    print(f"{table_num}번 테이블 청소완료!")
    print(f"{table_num}번 테이블 청소완료!")
    print(f"{table_num}번 테이블 청소완료!")
    print(f"{table_num}번 테이블 청소완료!")
```
