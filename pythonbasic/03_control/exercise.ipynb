# 01. 조건문



### 조건문이란

<aside>
💡 조건에 따라 수행할 코드를 다르게 작성하는 방법으로, 파이썬 키워드를 사용하여 작성한다. 조건문 작성에는 조건식이 필요하며, 조건식은 True 혹은 False의 결과를 내는 구문을 의미한다.

</aside>

```python
my_teacher = 'capybara'

if my_teacher == 'capybara':
    print("우리 썜은 카피바라 쌤입니다.")

if my_teacher == 'tiger':
    print("우리 썜은 호랑이 쌤입니다.")
```

```python
my_teacher = 'capybara'

if my_teacher == 'capybara':
    print("우리 썜은 카피바라 쌤입니다.")

else:
    print("우리 썜은 그럼 누구?")
```

```python
my_teacher = 'monkey'

if my_teacher == 'tiger':
    print("우리 썜은 호랑이 쌤입니다.")
elif my_teacher == 'monkey':
    print("우리 썜은 원숭이")
else:
    print("우리 썜은 카피바라")
```

### *조건문 match

```python
print("=== OhGiraffers vending machine ===")
print("tiger: 우리는 1인 24시간 교대제로 간다")
print("monkey: 24시간 문의 환영, 다른 말로 하면?")
print("squirrel: 저도 하루에 너다섯 시간 자는데요?")
print("capybara: 살아남는 사람이 강한 사람입니다.")
print("====================================")
print("만나고 싶은 선생님을 입력해 주세요")
input_teacher = input()
print("====================================")

sleep_time = 24
match input_teacher:
    case "tiger":
        print("호랑이 선생님의 메타버스에 오신 것을 환영합니다.")
        sleep_time = 3
    case "monkey":
        print("원숭이 선생님의 밀림에 오신 것을 환영합니다.")
        sleep_time = 4
    case "squirrel":
        print("다람쥐 선생님의 매력의 늪에 오신 것을 환영합니다.")
        sleep_time = 5
    case "capybara":
        print("카피바라 선생님의 교실에 오신 것을 환영합니다.")
        sleep_time = 6
    case _:
        print("공부할 마음이 없으신 건 아니죠??")
     
print(f"매일 {sleep_time} 시간의 수면이 예상됩니다.")
```

```python
lst = [n * n for n in range(1,11) if n % 2 !=0]
print("홀수 제곱근 : ", lst,end='\n\n')
lst = [n**2 for n in range(1,11) if n % 2 == 0]
print("짝수 제곱근 : ", lst)
```

### 03-01. break


```python
for x in [1, 2, 3]:
    
    print("앞", x)
    if x == 2:
        print("if", x)
        print("끝")  	
        break
    
    print("뒤", x)
print("진짜 끝!")
```

```python
for x in [3,4,5,6,7]:
    if x == 5:
        print("찾았다",x)
        break
        # continue
    print("못찾았다")
```

```python
for i in range(3):
    for j in range(3):
        if j == 1:
            break
        print("i = ", i ,"j =", j)
```

```python
while True:
    print("실행")
    break
```

### 03-02. continue


```python
for x in range(3):	
    print("앞", x)
    
    if x == 1:
        print("if", x)
        print("continue 등장!")  	
        continue
    
    print("뒤", x)
```

```python
for num in range(1,6):
    if num % 2 == 0:
        continue
    print(num)
```
