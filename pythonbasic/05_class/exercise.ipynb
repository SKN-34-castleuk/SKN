# 01. 클래스 (Class)

## 01-01. 클래스 개요

### 01-01-01. Class란

<aside>

## 01-02. 클래스 구성 요소

### 01-02-01. 클래스 속성

- 클래스 자체에 속하는 변수로, 모든 인스턴스가 공유하는 속성이다.

```python
class Person:
    national = 'korea'
    language = 'korean'
    

p1 = Person()
p2 = Person()

print(p1.national)  # korea
print(p2.national)  # korea
print(p1.language)  # korean
print(p2.language)  # korean
```

### 01-02-02. 메서드 (method)

- 클래스 내부에 정의된 함수로, 인스턴스의 데이터를 조작하거나 동작을 정의한다.
- self
    - 메서드 내에서 쓰이는 self는 필드 및 메소드에 접근하기 위한 객체를 의미한다.
    - 메소드 호출 시 객체의 주소값이 첫 번째 인자로 넘어오기 때문에, 객체를 통한 접근 시 호출되는 메소드의 첫 번째 인자는 항상 self 여야 한다.

```python
class Person:
    national = 'korea'
    language = 'korean'

    def greeting(self):
        return '안녕하세요'
    
    def information(self):
        return "I'm from" + self.national + " and i use " + self.language
    
    def favorite(self, color):
        return " i Love " + color
    

p1 = Person()
p2 = Person()

print(p1.favorite('red'))
print(p2.favorite('blue'))
```

### 01-02-03. 생성자

- __init__ 메서드는 객체가 생성될 때 자동으로 호출되는 메서드로, 생성자라고 부른다. 이때 매개변수를 전달받아 인스턴스 속성을 초기화 할 수 있다.
- 인스턴스 속성이란 각 인스턴스마다 개별적으로 가지는 변수이며, 생성자에서 정의된다.

```python
class Person:
    national = 'korea'
    language = 'korean'

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def greeting(self):
        return '안녕하세요'
    
    def information(self):
        return "I'm from " + self.national + " and I use " + self.language + ". My name is " + self.name + ". I'm " + str(self.age)
    
    def favorite(self, color):
        return "I love " + color 
    
p1 = Person('최성욱',30)
print(p1.national)  
print(p1.language)  
print(p1.name)
print(p1.age)
print(p1.information())
print(p1.favorite("black"))



```

## 01-03. 클래스 심화

### 01-03-01. **네임 스페이스와 스코프**

- 변수는 정의된 위치에 따라 **사용 가능한 범위(스코프)** 가 결정된다.
- 파이썬은 변수를 찾을 때 **가까운 범위부터 순서대로 검색**한다.
- 핵심 규칙 (LEGB)
    
    > 변수는 아래 순서로 찾는다.
    > 
    1. Local (현재 함수 내부)
    2. Enclosing (바깥 함수)
    3. Global (전역)
    4. Built-in (파이썬 내장)

```python
x = "global"

def outer():
    x = "outer"
    
    def inner():
        x = "inner"
        print(x)
    
    inner()

outer()    # inner
```

```python
class Test:
    x = "class"
    
    def show(self):
        print(self.x)

obj = Test()
obj.x = "instance"
obj.show()    # instance
```

### 01-03-02. **global과 nonlocal**

- global
    - 함수 내부에서 **전역 변수 값을 변경**할 때 사용한다.

```python
g_variable = "global variable"

def modify_global():
    global g_variable
    g_variable = "global modified"

print(g_variable)    # global variable
modify_global()
print(g_variable)    # global modified
```

- nonlocal
    - 중첩 함수에서 **바깥 함수의 변수 값을 변경**할 때 사용한다.

```python
def outer_function():
    variable = "outer variable"

    def inner_function():
        nonlocal variable
        variable = "modified"

    print(variable)
    inner_function()
    print(variable)

outer_function()
```

### 01-03-03. **Private Variable**

- 파이썬은 완전한 의미의 `private` 속성을 강제하지 않는다.
- 대신, 변수 이름 앞에 `__`(언더바 2개)를 붙여 **외부에서 직접 접근하지 않도록 권장하는 방식**을 사용한다.

```python
class Character:
    def __init__(self, nickname, type):
        self.nickname = nickname

        self.__type = type

    def get_type(self):
        return self.__type
    
    def set_type(self,type):
        character_types = ['전사','법사','치유사']
        if type in character_types:
            self.__type = type
        
        else:
            raise ValueError("잘못된 캐릭터 타입입니다.")
        
my_character = Character("카피바라GO","전사")

print(my_character.nickname)
# print(my_character.__type)


try:
    print(my_character.__type)

except AttributeError as e:
    print(e)


print(my_character.get_type())


my_character.set_type("법사")
print(my_character.get_type())
# my_character.set_type("흑")


try:
    my_character.set_type("흑마법사")

except ValueError as e:
    print(e)
```

## 02-01. 인스턴스 개요

### 02-01-01. 인스턴스란

### 02-01-02. 인스턴스 생성

- 클래스명 뒤에 `소괄호 ()`를 붙여 객체를 생성한다.
    - 클래스 내에 생성자(__init__ 메서드)가 정의되어 있는 경우, 생성자에서 사용하는 매개변수를 전달하면서 객체를 생성할 수 있다.
    - 같은 클래스로 여러 개의 객체를 만들 수 있다.

```python
class Person:
    national = 'korea'
    language = 'korean'

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def greeting(self):
        return '안녕하세요'
    
    def information(self):
        return "I'm from " + self.national + " and I use " + self.language + ". My name is " + self.name + ". I'm " + str(self.age)
    
    def favorite(self, color):
        return "I love " + color 
    
squirrel = Person('다람쥐', 900)
capybara = Person('카피바라', 20)

print(squirrel)    # <__main__.Person object at 0x000002E16EE94560>
print(capybara)     # <__main__.Person object at 0x000002E16EE94BF0>

print(squirrel.national)            # korea
print(squirrel.greeting())          # 안녕하세요
print(capybara.information())        # I'm from korea and I use korean. My name is 카피바라. I'm 20
print(capybara.favorite('black'))    # I love black

print(squirrel.name)  
print(capybara.name) 
```

## 02-02. 인스턴스 속성

### 02-02-01. 클래스 속성과 인스턴스 속성

- 인스턴스 변수는 인스턴스별 데이터를 위한 것이고, 클래스 변수는 그 클래스의 모든 인스턴스에서 공유되는 어트리뷰트와 메서드를 위해 사용한다.
    - 인스턴스 속성은 객체마다 개별 데이터를 의미한다.
    - 클래스 속성은 모든 객체가 공유되는 속성으로 변경이 발생하면, 전체 객체에서 변경될 수 있다.
    - 리스트, 딕셔너리는 반드시 인스턴스 속성으로 만들어야 한다.

```python
class Character:    
    skills = []
    
    def __init__(self, nickname, type):
        self.nickname = nickname
        self.type = type
        
    def add_skill(self, skill):
        self.skills.append(skill)
    
    def show_skill(self):
        return self.type + ' ' + self.nickname + '님의 보유 스킬 ' + str(self.skills)

my_character = Character('산골짜기다람쥐', '전사')
your_character = Character('도시카피바라', '법사')
my_character.add_skill('곤봉 휘두르기')
your_character.add_skill('마법진 그려서 날려버리기')

print(my_character.show_skill())    # 전사 산골짜기다람쥐님의 보유 스킬 ['곤봉 휘두르기', '마법진 그려서 날려버리기']
```

- 클래스 속성에 리스트를 두면 모든 객체가 같은 리스트를 공유하게 된다.
- 따라서 의도하지 않은 데이터 공유 문제가 발생할 수 있기 때문에 객체마다 다른 데이터를 관리해야 하는 경우에는 인스턴스 속성으로 설정해야 한다.

```python
class Character:        
    def __init__(self, nickname, type):
        self.nickname = nickname
        self.type = type
        self.skills = []
        
    def add_skill(self, skill):
        self.skills.append(skill)
    
    def show_skill(self):
        return self.type + ' ' + self.nickname + '님의 보유 스킬 ' + str(self.skills)

my_character = Character('산골짜기다람쥐', '전사')
your_character = Character('도시카피바라', '법사')
my_character.add_skill('곤봉 휘두르기')
your_character.add_skill('마법진 그려서 날려버리기')

print(my_character.show_skill())    # 전사 산골짜기다람쥐님의 보유 스킬 ['곤봉 휘두르기']
```

- 인스턴스 속성과 클래스 속성으로 같은 이름을 사용하면, 인스턴스 속성을 우선한다.


```python
class Character:        
    nickname = '땅파먹는개미핥기'
    
    def __init__(self, nickname, type):
        self.nickname = nickname
        self.type = type

my_character = Character('도시카피바라', '법사')

print(my_character.nickname)    
```

03-01-03. 상속 기본 문법

class DerivedClassName(BaseClassName):
    <statement-1>
    .
    .
    .
    <statement-N>

* Student는 Person의 기능을 그대로 사용할 수 있다.

```python
class Person:
    national = 'korea'

    def greenting(self):
        return 'Hello. This is Python'
    
class Student(Person):
    pass


student = Student()
print(student.greenting())
print(student.national)
```

```python
class Person:
    def __init__(self,name):
        self.name = name

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)
        # parent().__init(name)
        self.grade = grade


student = Student("oL","5학년")
print(student.name)
print(student.grade)
```

## 03-02. 상속의 특징

### 03-02-01. **다중 상속 지원**

- 부모를 여러 개 가지는 다중 상속을 지원한다.

class DerivedClassName(Base1, Base2, Base3):
    <statement-1>
    .

- 다중 상속 시  시 검색 순서는 왼쪽에서 오른쪽으로 탐색한다.

```python
class Person:
    national = 'korea'
    def greenting(self):
        return 'Hello. This is Python'
    
class Leaner:
    def greeting(self):
        return 'Hello. I am Learner'    
    def learn(self):
        return 'I am learning Python'
    
class Student(Person,Leaner):
    pass

student = Student()

print(student.greenting())
print(student.learn())

```

### 03-02-03. 부모 클래스의 속성 오버라이딩 가능

- 오버라이딩: 부모 클래스의 메서드를 자식 클래스에서 다시 정의하는 것
- 부모 클래스의 속성을 오버라이딩하여 사용할 수 있다.
- 같은 이름의 메서드를 만들면 부모의 것이 아니라 자식 것이 실행된다.

```python
class Person:
    national = 'korea'
    def greenting(self):
        return 'Hello. This is Python'
    
class Learner(Person):
    def __init__(self, subject):
        self.subject = subject
        
    def learn(self):
        return f"I am learning {self.subject}"
    
class Student(Learner):
    def __init__(self,name,subject):
        super().__init__(subject)
        self.name = name
    
    def greeting(self):
        return f'Hello. My name is {self.name}'


student = Student('카피바라', 'python')

print(student.greenting())

print(student.learn())

print(student.national)
```

```python
class Person:
    def greeting(self):
        return '안녕하쇼'
    

class Student(Person):
    def greeting(self):
        return super().greeting() + " 난 학생이오"
    
student = Student()
print(student.greeting())
```

```python
class Person:
    national = 'korea'

class Student(Person):
    national = "USA"
person = Person()
student = Student()


print(person.national)
print(student.national)
    
```

super()를 사용하면 부모 클래스의 메서드를 호출할 수 있다.

```python
class Animal:
    def do(self):
        print('👉👈')
class Dog(Animal):
    def do(self):
        print('🐶🐶🐶')
class Duck(Animal):
    def do(self):
        print('🦆🦆🦆')
class Fish(Animal):
    pass    # do 호출시 부모 클래스 Animal#do가 호출된다.

for each in (Dog(), Duck(), Fish()):
    each.do()
```

```python
# 상속 관계가 아니더라도 동일한 이름의 함수가 있는 경우 다형성처럼 사용이 가능
class Alarm:
    def do(self):
        print('⏰⏰⏰')

for each in (Dog(), Duck(), Fish(), Alarm()):
    each.do()
```

# enum 클래스

```python
from enum import Enum

class Gender(Enum):
    M = '남'
    F = '여'

print(Gender.M)
print(Gender.F)

# name
print(Gender.M.name)
print(Gender.F.name)

# value
print(Gender.M.value)
print(Gender.F.value)
```

```python
print(Gender.M == 'M') # False
print(Gender.M.name == 'M') # True

a = Gender.M
b = Gender.M

print(a == b)  # True
print(a is b)  # True
```

```python
Color = Enum('Color', ['RED', 'GREEN', 'BLUE'])
for color in Color:
    print(color.name, color.value) 
```

```python
from enum import Enum, auto

class Lang(Enum):
    C = auto()        # 1
    Java = auto()     # 2
    Python = auto()   # 3

for lang in Lang:
    print(lang.name, lang.value)
```

```python
from enum import Enum

class Number(Enum):
    ONE = 1
    FIRST = 1   # ONE의 별칭(alias)

print(Number.ONE)    # Number.ONE
print(Number.FIRST)  # Number.ONE
print(Number.ONE == Number.FIRST)   # True
print(Number.FIRST is Number.ONE)   # True

for num in Number:
    print(num)   # 대표 멤버만 순회
```
