### **문제 2: `.csv` 파일 생성 및 읽기**

카피바라는 이직 준비를 위해 지원했던 회사들의 정보를 저장하려고 합니다.

1. `applications.csv` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```
    회사명,직무,결과
    회사A,데이터 분석가,합격
    회사B,소프트웨어 엔지니어,불합격
    회사C,제품 관리자,진행 중
    ```
    
2. 파일을 닫고 다시 열어 모든 데이터를 출력하세요.

````python
import csv

data = [
    ["회사명","직무","결과"],
    ["회사A","데이터 분석가","합격"],
    ["회사B","소프트웨어 엔지니어","불합격"],
    ["회사C","제품 관리자","진행 중"]
]


with open("appliactions.csv","w",encoding="utf-8") as f:    
    write = csv.writer(f)  # CSV 작성도구
    write.writerows(data)  # 열,행을 한꺼번에 작성

with open("appliactions.csv","r",encoding="utf-8") as f:
    data = csv.reader(f)  # 한 줄씩 리스트 형태로 생성 
    for row in data:
        print(", ".join(row))
````

### **문제 4: `.json` 파일 생성 및 읽기**

카피바라는 좋아하는 음식 데이터를 저장하고 싶습니다.

1. `favorite_foods.json` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```json
    {
        "name": "카피바라",
        "favorite_foods": ["도토리", "호두", "밤"]
    }
    ```
    
2. 파일을 읽어와 카피바라의 좋아하는 음식 리스트를 출력하세요.
- hint
    
    읽을 때 json 라이브러리가 필요하다.
    
    open을 한 json 파일 객체에서 데이터를 읽어올 때는 `json.load(파일 객체)`
    
    그 중에 favorite_foods만 가져오고 싶다면 인덱스값을 `읽어온 데이터 변수[”favorite_foods”]`

````python
import json

data = {
    "name": "카피바라",
    "favorite_foods": ["도토리", "호두", "밤"]
}

with open("favorite_foods.json","w",encoding="utf-8") as f:
    json.dump(data,f, ensure_ascii=False, indent=4)

with open("favorite_foods.json","r",encoding="utf-8") as f:
    json_data = json.load(f)
    # print(json_data)
    print(", ".join(json_data["favorite_foods"]))
````

````python
import json
data = {
    "name": "카피바라",
    "favorite_foods": ["도토리", "호두", "밤"]
}

with open("favorite_foods.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)


with open("favorite_foods.json", "r", encoding="utf-8") as f:

    json_data = json.load(f)
    print(json_data["favorite_foods"])
````

### **문제 5: `.html` 파일 수정**

카피바라는 자신의 이력서를 작성 중입니다. `resume.html` 파일을 다음 내용으로 생성하고 수정하세요.

1. `resume.html` 파일 생성 내용:
    
    ```html
    <html>
        <head><title>카피바라의 이력서</title></head>
        <body>
            <h1>카피바라의 이력서</h1>
            <p>경력: 신입</p>
        </body>
    </html>
    ```
    
2. 수정 작업:
    - `<p>` 태그 안의 "신입"을 "1년 경력"으로 변경하세요.
3. 수정된 파일을 읽어와 전체 내용을 출력하세요.

````python
data = """
<html>
    <head><title>카피바라의 이력서</title></head>
    <body>
        <h1>카피바라의 이력서</h1>
        <p>경력: 신입</p>
    </body>
</html>
"""


with open("resume.html", "w", encoding="utf-8") as f:
    f.write(data)
with open("resume.html", "r", encoding="utf-8") as f:
    data =f.read()
    print(data)
````
