## 01-01. 모듈 개요

### 01-01-01. 모듈이란

<aside>
💡 하나의 모듈은 서로 관련된 코드의 집합이자 하나의 파일을 의미하며, 함수, 클래스, 변수 등을 모듈 내에 정의할 수 있다. 이렇게 만든 모듈은 다른 Python 프로그램에서 가져와 사용할 수 있다. 

정리하자면 다른 파일에 있는 코드를 가져다 쓰는 것이다.

</aside>

```python
import ohgiraffers_module

print(ohgiraffers_module.ohgiraffers)    # 국내최고 IT 전문교육 강사팀
print(ohgiraffers_module.gorilla())      # 크왕
print(ohgiraffers_module.pig())          # 꿀꿀
print(ohgiraffers_module.tiger())        # 어흥
```

```python
from ohgiraffers_module import ohgiraffers, gorilla

print(ohgiraffers)    # 국내최고 IT 전문교육 강사팀
print(gorilla())      # 크왕
```

```python
import ohgiraffers_module as OGModule

print(OGModule.ohgiraffers)    # 국내최고 IT 전문교육 강사팀
print(OGModule.gorilla())      # 크왕
```

```python
from ohgiraffers_module import ohgiraffers as OG, gorilla as grl

print(OG)       # 국내최고 IT 전문교육 강사팀
print(grl())    # 크왕
```

표준 라이브러리 모듈을 import 하여 사용할 수도 있다.
    
    
    - Python은 기본적으로 다양한 모듈을 제공하므로 별도 설치 없이 사용할 수 있다.

```python
# math 모듈 예시
import math

print(math.sqrt(16))        # 출력: 4.0
print(math.factorial(5))    # 출력: 120
print(math.pi)              # 출력: 3.141592653589793
```
