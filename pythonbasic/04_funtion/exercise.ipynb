## 01-01. 매개변수

### 01-01-01 파라미터

- 함수 생성 시 함수명 옆에 붙이는 소괄호 안에 매개변수로 전달받을 값의 변수명을 넣을 수 있다. 이를 파라미터라고 한다.

```python
def greetiong(name,greet):
    print(name,":",greet)

greetiong("다람쥐","감사합니다람쥐~")
```

```python
def summ(a,b):
    print(f"{a} + {b} = {a+b}")

summ(1,3)
```

- 함수를 정의할 때 매개변수의 기본값을 지정해 줄 수 있다.
    
    함수 호출 시 인자를 보내지 않으면 함수 선언부에서 미리 지정한 기본값을 사용하여 함수를 실행한다.

```python
def greeting(name, greet="처음 뵙겠습니다!!!"):
    print(name, ":", greet)

greeting("다람쥐")    
```

### 01-02-02. 인자

- 함수 호출 시 함수명에 소괄호를 붙여 호출하며, 이때 소괄호 안에 전달하는 값은 인자라고 한다.
- 함수 인자를 보내는 방식은 다음 2가지가 있다.
    1. 위치 인자: 위치로 매칭하는 방법
    2. 키워드 인자: 매개변수 이름으로 매칭하는 방법
    - 이때, 위치 인자를 먼저 사용하면 뒤에 키워드 인자를 쓸 수 있다.
    - 단, 위치 인자와 키워드 인자를 함께 사용할 때 키워드 인자를 먼저 사용하는 것은 불가하다.

```python
def information(name, color):
    print(name + '의 최애 색상', color,sep = ' : ')

information('capybara','black')
```

```python
information(color="black",name = "capybara")
information('capybara', color = "black")
# information(color='black', 'monkey')    # 불가능
# information(name='monkey', 'black')     # 불가능
```

## 01-02. 반환값

### 01-02-01. return

- 함수 블럭 내부에서 return 키워드를 만나면 함수의 작동이 중단된다.
- return 키워드 뒤에 변수 또는 값이 오면, 해당 자료를 함수 호출 위치에 반환하며 중단된다.

```python
def introduce(name):
    return f"안녕하세요 {name} 강사 인사드립니다~!"

result = introduce("카피바라")
print(result)
```

- return 키워드는 선택적으로 사용할 수 있다. (반드시 작성하지 않아도 된다.)

```python
def test():
    print("hello")
    # return "hello"

result = test()
print(result)
```

- return이 없으면 함수 실행만 진행되고, 반환값을 받을 시 None이 반환된다.
- return vs print()
    - return은 값을 “돌려준다”
    - print는 화면에 “출력만 한다”

```python
def calc(a, b):
    return a+b,a-b

print(calc(2,3))
print(type(calc(2,3)))
```

```python
plus,minus = calc(2,3)

print(f"plus 는 {plus} minus 는 {minus}")
```

## 01-03. 패키지와 언패키지

### 01-03-01. 패키징

- 위치 인자가 많을 때 *을 붙여서 패키징하여 하나의 객체로 처리할 수 있다.
- 매개변수 개수가 정해지지 않았을 때 사용한다.

```python
def add_many(*args):
    result = 0

    for i in args:
        result += i

    return result

print(add_many(1,2,3,4,5))
```

- 키워드 인자인 경우에는 * 두 개(**)를 붙여서 작성할 수 있다.
- `*args` 는 여러 값을 **튜플로 묶어 받고**, `**kwargs` 는 이름이 있는 값을 **딕셔너리로 묶어 받는다.**
- `args`, `kwargs`는 문법이 아닌 관례적인 이름이며, 실제로는 `*`와 `**` 기호가 기능을 결정한다.

```python
def show(*args):
    print(args)

show(1,2,3,"capybara")
```

```python
def show_info(**kwargs):
    return kwargs
    
print(show_info(name = "capybara", age = 30))
```

### 01-03-02. 언패키징

- 패키지와 반대되는 개념으로 여러 개의 객체를 포함하는 하나의 객체를 풀어준다.

```python
def sum(a,b,c):
    return a + b + c

numbers = [1, 2, 3]
# sum(numbers) #에러 발생
print(sum(*numbers))
```

- 리스트를 풀어서 전달하지만 내부적으로 함수 안에서는 다시 **tuple로 묶임**

## 01-04. 람다 (lambda)

### 01-04-01. 람다란

<aside>
💡 람다는 일회성의 간단한 함수를 정의할 때 유용하게 사용할 수 있는, 함수를 보다 간단하고 쉽게 선언하는 방법이다. 매개변수로 함수를 전달해야 할 때, 함수 구문을 작성하는 것이 번거롭고 코드가 지나치게 길어지는 경우 사용한다. 즉, 함수 기능을 매개변수로 전달하는 코드를 더 효율적으로 작성할 수 있도록 한다.

</aside>

1. 람다 기본

```python
lambda x : x+1
lambda x : x**2
lambda x : x[1]


lambda x:(x+1 if x >0 else x-1)
```

```python
add = lambda x, y : x + y
print(add(3,5))
```

```python
def add(x,y):
    return x+y

print(add(3,5))
```

2. 람다 합수를 함수의 인자로 전달

```python
numbers = [1,2,3,4,5]
squared_numbers = map(lambda x: x**2,numbers)

# print(squared_numbers)
print(list(squared_numbers))
```

3. 람다 함수를 고차함수의 인자로 전달 (배열 요소 필터링)

```python
numbers = [1,2,3,4,5,6]
even_numbers = filter(lambda x: x%2==0,numbers)

print(list(even_numbers))
```

4. 배열 정렬 기준을 람다 함수로 전달

```python
points =[(1, 2), (3, 1), (5, -1)]
sorted_points = sorted(points,key=lambda x: x[1])
print(sorted_points)
```

5. 객체(딕셔너리) 기준 정렬

```python
students = [
    {"name":"bear", "score":90},
    {"name":"tiger", "score":70},
    {"name":"rabbit", "score":85}
]

print(sorted(students, key=lambda x: x["score"]))
```

### 함수의 종류

- **파라미터와 반환값에 따른 종류**
    1. param, no return value
    2. param, return value
    3. no param, no return value
    4. no param, return value

```python
#immutable 객체 (변경 불가능) : int, str, tuple 등

def f(x):
    x = 100

a = 10
f(a)
print(a)    # 10
```

```python
#mutable 객체 (변경 가능) : list, dict 등

def f(x):
    x[0] = 100

a = [1, 2, 3]
f(a)
print(a)    # [100, 2, 3]
```

## 1️⃣ zip & unzip

> 
> 
> 
> 여러 개의 **iterable(리스트, 튜플 등)** 을 동일한 인덱스끼리 **쌍(pair)** 으로 묶어주는 함수.
> 
> https://www.daleseo.com/python-zip/
> 
- zip은 같은 인덱스끼리 묶는다.
- 길이가 다르면 짧은 쪽 기준으로 묶인다. (길이가 다르면 긴 쪽은 버려짐)
- 두 리스트를 동시에 순회할 때 자주 사용한다.


```python
letters = ["A", "B", "C"]
numbers = [1,2,3]
for pair in zip(letters, numbers):
    print(pair)
```

```python
for x,y in zip(letters,numbers):
    print(x,y)
```

```python
dict(zip(letters, numbers))
```

## unzip(언패킹)

```python
pairs = [('A', 1), ('B', 2), ('C', 3)]

letters, numbers = zip(*pairs)

print(letters)
print(numbers)
```

```python


print(zip(letters, numbers))        
print(list(zip(letters, numbers)))
```

## 2️⃣ sort & sorted

```python
#`sort()`는 원본을 바꾸고 반환값이 없다.

lst = [3, 1, 2]
lst.sort()
print(lst.sort())  # None
print(lst)   # [1, 2, 3]


# sort()`는 반환값이 `None`이므로 `print`하면 `None`이 나온다.
```

- `sorted()`는 원본 유지 + 새로운 리스트를 반환한다.


```python
lst = [3, 1, 2]
new_lst = sorted(lst)
print(lst)      # [3, 1, 2]
print(new_lst)  # [1, 2, 3]
```

```python
## dict 정렬
d = {'blockdmask': 400, 'equal': 300, 'apple': 200, 'dish': 100, 'cook': 500}

print(d)
print()
print(d.keys())
print()
print(d.items())
print()


# value 기준
print(sorted(d.items(), key = lambda x : x[1]))
print()
# key 기준
print(sorted(d.items(), key = lambda x: x[0]))
```

## **3️⃣ dir**

- `dir()` 함수는 파이썬에서 객체가 가진 속성과 메서드를 나열해주는 함수입니다. 객체를 인수로 전달하지 않으면 현재의 전역 심볼 테이블에 정의된 이름들을 반환하고, 특정 객체를 인수로 전달하면 그 객체의 속성 목록을 반환합니다.


1. 인수를 전달하지 않는 경우 현재 스코프에서 사용할 수 있는 모든 변수와 메서드 등을 반환

```python
dir()
['__annotations__',
'__builtins__',
'__doc__', 
'__name__', 
'__package__']
```

2. 인수를 전달하는 경우 전달한 객체가 어떤 속성과 메서드를 가지고 있는지 알려줌
    
    
    - 예를 들어 `dir(str)`는 `str` 클래스가 가지고 있는 모든 메서드와 속성을 나열

```python
dir(str)
['__add__',
'__class__',
'__contains__',
'__delattr__',
'__dir__',
...]
```

- 객체가 어떤 메서드나 속성을 가지고 있는지 탐색할 때 유용
- 특정 모듈이나 클래스에 대한 내용을 빠르게 확인 가능
