```python
import numpy as np

arr = np.array([30,10,20,40,50])
sorted_arr = np.sort(arr)
print(sorted_arr)
sorted_arr_desc = sorted_arr[::-1]
print(sorted_arr_desc)
```

```python
arr2d = np.array([[5, 2, 3], [8, 4, 1]])
sorted_arr2d = np.sort(arr2d)
print(sorted_arr2d)
```

```python
np.sort(arr2d,axis=1)
```

```python
arr = np.array([30, 10, 20])
sorted_indices = np.argsort(arr)
print(sorted_indices)
print(arr[sorted_indices])
```

```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
merged_arr = np.concatenate((arr1, arr2))
print(merged_arr)  # [1, 2, 3, 4, 5, 6]
```

```python
arr1 = np.array([
    [1, 2], 
    [3, 4]
])
arr2 = np.array([
    [5, 6], 
    [7, 8]
])
merged_arr = np.concatenate((arr1,arr2),axis=0)
print(merged_arr,merged_arr.shape)
```

```python
merged_arr = np.concatenate((arr1,arr2),axis=1)
print(merged_arr,merged_arr.shape)
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
merged_arr = np.vstack((arr1, arr2))
print(merged_arr)
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
merged_arr = np.hstack((arr1, arr2))
print(merged_arr) 
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
concat__arr1 = np.concatenate((arr1,arr2))
print("concat_arr = ",concat__arr1)
concat__arr2 = np.concatenate((arr1,arr2))
print("concat_arr = ",concat__arr2)

print()
vs_arr = np.vstack((arr1,arr2))
print("vs_arr = ",vs_arr,vs_arr.shape)
print()
hs_arr = np.hstack((arr1,arr2))
print("hs_arr = ",hs_arr)
print()
stacked_arr = np.stack((arr1, arr2), axis=0)
print("stacked_arr = ",stacked_arr,stacked_arr.shape)
print()
stacked_arr1 = np.stack((arr1, arr2), axis=1)
print("stacked_arr = ",stacked_arr1,stacked_arr1.shape)
```

```python
import time
size = 1000000
python_list = list(range(size))

start_time = time.time()
python_list = [x * 2 for x in python_list]
print("Python 리스트 성능:", time.time() - start_time)
```

```python
import numpy as np
np_array = np.arange(size)

start_time = time.time()
np_array = np_array * 2
print("NumPy 배열 성능:", time.time() - start_time)
```
