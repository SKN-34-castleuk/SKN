- 선 그래프는 시간의 흐름이 들어간 데이터에 사용하면 좋음

```python
import matplotlib.pyplot as plt


months = ['Jan','Feb','Mar','Apr','May']
sales = [150,200,180,220,178]
plt.plot(months,
         sales,
         marker = '^',
         color = 'blue')
plt.bar(months,sales,color = 'skyblue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
for i, month in enumerate(months):
    plt.text(month,sales[i]+ 5,sales[i])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
np.random.seed(0)
study_hours = np.random.normal(5, 1.5, 100)  # 평균 5시간, 표준편차 1.5시간
scores = study_hours * 10 + np.random.normal(0, 5, 100)  # 공부 시간에 따른 점수

plt.scatter(study_hours,
            scores,
            color="purple", 
            alpha=0.6)
plt.title('Relationship between Study Hours and Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

```python
import matplotlib.pyplot as plt

# X축과 Y축 데이터 설정
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [30, 25, 20, 10]
y3 = [20, 30, 10, 40]
y4 = [10,5,25,40]

# 다양한 선 스타일로 그래프 그리기
plt.plot(x, y1, linestyle='-', label='Solid Line')
plt.plot(x, y2, linestyle='--', label='Dashed Line')
plt.plot(x, y3, linestyle=':', label='Dotted Line')
plt.plot(x, y4, linestyle='-.', label='Dash_Dot Line')

# 레이블 및 타이틀 추가
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Styles Example')

# 범례 추가
plt.legend()

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt

# X축 데이터 설정
x = [1, 2, 3, 4, 5]

# Y축 데이터 설정
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

# 첫 번째 선: 빨간 실선, 투명도 설정
plt.plot(x, y1, label='y = x^2', color='r', alpha=0.3)

# 두 번째 선: 파란 점선, 투명도 설정
plt.plot(x, y2, label='y = -x^2', color='b', alpha=0.8,linestyle = '--')

# 레이블 및 타이틀 추가
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Lines with Transparency')

# 범례 추가
plt.legend()

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)

# 서브플롯 생성
plt.figure(figsize=(10, 5))

# 첫 번째 그래프
plt.subplot(1, 2, 1)   # 1행 2열 중 첫번째
plt.plot(x, np.sin(x))
plt.title('Sine Wave')

# 두 번째 그래프
plt.subplot(1, 2, 2)   # 1행 2열 중 두번째
plt.plot(x, np.cos(x))
plt.title('Cosine Wave')

plt.tight_layout()       # 레이아웃 자동조정
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 2D 배열 데이터 생성
data = np.random.rand(10, 10)

# 두 개의 서브플롯에서 다른 컬러맵 적용
plt.subplot(1, 2, 1)
plt.imshow(data, cmap='plasma')
plt.title('Plasma Colormap')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(data, cmap='inferno')
plt.title('Inferno Colormap')
plt.colorbar()

# 서브플롯 간격 조정
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# 사용자 정의 컬러맵 설정
# LinearSegmentedColormap.from_list()를 사용하여 색상 리스트를 기반으로 컬러맵 정의
colors = ["#060606", "#53CBFF", "#FFFFFF"]  # 파란색에서 녹색, 빨간색으로 변환
custom_cmap = LinearSegmentedColormap.from_list('custom', colors)

# 2D 배열 데이터 생성
data = np.random.rand(10, 10)

# 사용자 정의 컬러맵 적용
plt.imshow(data, cmap=custom_cmap)
plt.colorbar()

# 타이틀 추가
plt.title('Custom Colormap')

# 그래프 보여주기
plt.show()
```
