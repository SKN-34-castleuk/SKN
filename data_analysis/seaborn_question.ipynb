````python
import seaborn as sns
import matplotlib.pyplot as plt
# import matplotlib.font_manager as fm
titanic = sns.load_dataset('titanic')
#한글 폰트 설정 (mac은 AppleGothic)
plt.rcParams["font.family"] = 'Malgun Gothic'
# 마이너스 깨짐
plt.rcParams["axes.unicode_minus"] = False

titanic
````


- 문제 1: 생존자와 사망자 수 시각화
    - Titanic 데이터셋의 `survived` 열을 사용하여 생존자와 사망자의 수를 막대 그래프로 시각화하세요.
    - 그래프에 제목, 축 레이블을 추가하고 색상을 설정하세요.
    


````python
plt.Figure(figsize= (10,6))
sns.countplot(x = 'survived',data = titanic, palette= 'muted',hue='survived')
plt.title('생존자와 사망자 수')
plt.xlabel('생존여부(1 = 생존 , 0 = 사망)')
plt.ylabel('수')
plt.show()
````

- 문제 2: 성별에 따른 생존율 비교
    - Titanic 데이터셋의 `sex`와 `survived` 열을 사용하여 성별에 따른 생존율을 막대 그래프로 시각화하세요.
    - 성별 생존율을 비교하고 제목, 축 레이블을 추가하세요.
    


````python
titanic
````

````python
sns.barplot(x = 'sex',
            y = titanic['survived']*100,
            data = titanic,
            palette='coolwarm',
            hue = 'sex')
plt.xlabel('성별')
plt.ylabel('생존율(%)')
plt.title('성별에 따른 생존율')
plt.show()
````

- 문제 3: 클래스와 성별에 따른 생존율 비교
    - `class`를 x축으로 설정하고, 성별에 따라 색상을 구분하여 그래프를 생성하세요.
    - Titanic 데이터셋의 `class`, `sex`, `survived` 열을 사용하여 그룹별 생존율을 시각화하세요.
    - 그래프에 제목, 축 레이블, 범례를 추가하세요.
    


````python
titanic
````

````python
plt.Figure(figsize= (10,6))

sns.barplot(x = 'class',
            y=titanic['survived'] *100,
            data = titanic,
            palette= 'hot',
            hue= 'sex')
plt.xlabel('클래스')
plt.ylabel('생존율(%)')
plt.title('클래스와 성별에 따른 생존율')
plt.show()
````

- 문제 4: 나이 분포 분석
    - Titanic 데이터셋의 `age` 열을 사용하여 승객들의 나이 분포를 히스토그램과 KDE 그래프로 시각화하세요.
    - 데이터에서 결측치를 제거하고 그래프를 그리세요.
    - 히스토그램의 막대 개수(`bins`)를 적절히 설정하고 제목, 축 레이블을 추가하세요.
    


````python
titanic.isnull().sum()
````

````python
titanic['age'].dropna()
````

````python
plt.Figure(figsize= (10,6))

sns.histplot(titanic['age'].dropna(), kde=True, bins=25,color= 'skyblue')
plt.title('승객들의 나이분포')
plt.xlabel('나이')
plt.ylabel('승객수')
plt.show()
````

- 문제 5: 나이와 요금의 관계 분석
    - Titanic 데이터셋의 `age`, `fare`, `survived` 열을 사용하여 나이와 요금 간의 관계를 산점도로 시각화하세요.
    - 생존 여부(`survived`)에 따라 색상을 다르게 설정하세요.
    - 그래프에 제목, 축 레이블을 추가하세요.

````python
plt.Figure(figsize= (10,6))

sns.scatterplot(x=titanic['age'].dropna(), y='fare', data=titanic, hue='survived')

plt.title('승객들의 나이와 요금의 관계 분석')
plt.xlabel('나이')
plt.ylabel('요금')
plt.show()
````
