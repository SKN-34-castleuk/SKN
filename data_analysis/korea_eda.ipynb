```python
# t-test를 통해 성별 소득 차이의 통계적 유의성 검정

import scipy.stats as stats

sex_m = data[data['sex'] == 'M'].dropna()
sex_f = data[data['sex'] == 'F'].dropna()

stats.ttest_ind(sex_m['income'],sex_f['income'], equal_var=True)
```

t-test 결과 해석  
statistic(t-값) : 값이 크면 두 집단의 평균 차이가 크다.  
pvalue : 0에 거의 근접 -> p_value 기준인 0.05보다 작다. -> 귀무가설 기각  
남성과 여성의 평균 소득에는 유의미한 차이가 있다

```python
# age 컬럼생성 = 2026 - 출생연도

data['age'] = 2026 - data['birth']
# 나이를 정수형으로
data['age'] = data['age'].astype(int)
data
```

```python
#나이별 평균 소득

# 소득 결측치 제거
data_income = data.dropna(subset=['income'])

# 나이별 평균 소득 계산
data_income = data_income.groupby('age')['income'].mean()

# 선 그래프 시각화 (plot() : Series 형태로 선 그래프를 시각화)
data_income.plot()
```

```python
print(data_income.max())  # 나이별 가장 높은 평균소득
print(data_income.idxmax())  # 가장 소득이 높은 나이
```

```python
# 나이별 평균 소득을 내림차순 정렬

print(data_income.sort_values(ascending=False))
```

```python
# 새로운 파생변수 추가

data = data.assign(
    agetype = np.where(
        data['age'] < 30,'youth',   # 데이터 나이 30 미만 청년층
        np.where(
            data['age'] <60 ,'middle age','senior'  # 30~59 중장년층,60세 이상은 고령층
        )
    )
)
data
```

```python
# 연령대별 응답자수 
sns.countplot(x = 'agetype', data = data,hue='agetype')
```

```python
#연령대별 평균 소득
age_income = data.groupby('agetype')['income'].mean()
age_income
```

```python
# 소득 결측치 제거 후,(성별 & 나이)로 그룹화 하여 평균 소득 계산
# as_index =False : 그룹화한 컬럼을 인덱스로 쓰지않음

age_sex_income = data.dropna(subset=['income']).groupby(
    ['sex','age'], as_index= False    
    )['income'].mean()

age_sex_income
```

```python
sns.lineplot(x = 'age',
             y = 'income',
             data = age_sex_income, 
             hue= 'sex')
```

```python
# 남성/여성 소득 비율 그래프
# unstack() : 인덱스에 있던 값을 컬럼으로 펼쳐준다.
age_mean = data.dropna(subset=['income']).groupby(['age','sex'])['income'].mean().unstack()

age_mean['ratio_M_over_F'] = age_mean['M']/ age_mean['F']   # 남녀 소득비율 파생변수
age_mean
```

```python
# 연력별 소득비율 선그래프 시각화

age_mean['ratio_M_over_F'].plot(figsize=(8,3))
plt.axhline(1,linestyle = '--')
plt.title('연령대별 남여 소득비율')
plt.ylabel('M / F')
plt.show()
```

```python
job_name = pd.read_excel('data/(2025년 20차 한국복지패널조사) 조사설계서-가구용(beta1).xlsx',
                            sheet_name = '직종코드(2019 7차 신분류)'
                         )
job_name
```

```python
#소분류 -> job_code, Unnamed: 3 -> job, inplace = True

job_name.rename(columns={
        '소분류' : 'job_code',
        'Unnamed: 3' : 'job'
},inplace= True)

job_name = job_name[['job_code','job']]
job_name
```

```python
# data(분석데이터)와 job_name(직업별 매핑표)을 job_code 기준으로 매핑하여 결합 

job_df = data.merge(job_name, how = 'left', on='job_code')

job_df
```

```python
# 직업별 평균소득 계산

# 직업명 과 소득이 모두 존재하는 행만 필터링
job_income_mean = job_df.dropna(subset=['income','job'])
# 직업명을 기준으로 그룹화하여 퍙균 소득 계산
job_income_mean = job_income_mean.groupby(['job'])['income'].mean()

job_income_mean.sort_values()
```

```python
# 직업별 평균 소득을 내림차순으로 정렬하고 상위 10개만 출력

job_income_mean.sort_values(ascending=False).head(10)
```

```python
#평균 소득 상위 10개 직업 가로 막대 그래프로 시각화

top10 = job_income_mean.sort_values(ascending=False).head(10).sort_values()

plt.Figure(figsize= (8,6))
top10.plot(kind='barh')
plt.title("상위 10개 직업 평균 소득 랭킹")
plt.xlabel("평균소득(만원)")
plt.show()
```

```python
#평균 소득 하위 10개 직업 조회
low10 = job_income_mean.sort_values().head(10)
# low10 = job_income_mean.sort_values(ascending=False).tail()
low10
```

```python
#평균 소득 하위 10개 직업 가로 막대 그래프로 시각화

low10 = job_income_mean.sort_values().head(10).sort_values(ascending=False)

plt.Figure(figsize= (8,6))
low10.plot(kind='barh')
plt.title("하위 10개 직업 평균 소득 랭킹")
plt.xlabel("평균소득(만원)")
plt.show()
```

```python
#평균 소득 상위 10개 직업 가로 막대 그래프로 시각화

top20 = job_income_mean.sort_values(ascending=False).head(20).sort_values()

plt.Figure(figsize= (8,6))
top20.plot(kind='barh')
plt.title("상위 20개 직업 평균 소득 랭킹")
plt.xlabel("평균소득(만원)")
plt.show()
```
