## [실습 1] 다양한 그래프 활용 1

- 주어진 데이터에 대해 선 그래프, 막대 그래프, 히스토그램, 산점도를 한꺼번에 그려보세요.


````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

# 데이터 준비
x = np.arange(1, 11)
y = np.random.randint(10, 100, 10)

data = np.random.normal(50, 15, 100)
plt.figure(figsize=(10, 5))


# 선그래프
plt.subplot(1, 4, 1)   
plt.plot(x,y,marker='o')
plt.title('LineGraph')

# 막대그래프
plt.subplot(1, 4, 2)   
plt.bar(x,y,color = 'skyblue')
plt.title('BarGraph')

#히스토그램
plt.subplot(1, 4, 3)   
plt.hist(data,bins=10,color='green',edgecolor='black')
plt.title('Cosine Wave')

#산점도
plt.subplot(1, 4, 4)  
plt.scatter(x,y)
plt.title('Cosine Wave')


plt.tight_layout()      
plt.show()
````


## [실습 2] 다양한 그래프 활용 2

- **학생들의 시험 점수**
    - `math_scores`: [85, 90, 88, 76, 95, 89, 72, 84, 91, 87]

```jsx
months = ['January', 'February', 'March', 'April', 'May', 'June', 
            'July', 'August', 'September', 'October', 'November', 'December']
avg_temperatures = [-1, 0, 5, 10, 15, 20, 25, 23, 19, 12, 5, 0]
rainfall = [50, 40, 60, 80, 100, 120, 150, 140, 110, 90, 70, 60]

```

1. 월별 평균 기온을 시각화하는 선 그래프를 그려보세요.
2. 월별 강수량을 시각화하는 막대 그래프를 그려보세요.
3. 수학과 영어 점수의 분포를 비교하는 박스플롯 그래프를 그려보세요.
4. 수학 점수 분포를 분석하는 히스토그램을 그려보세요.
5. 수학 점수와 영어 점수의 상관관계를 분석하는 산점도를 그려보세요.
6. 1~5의 모든 그래프를 하나의 Figure에 배치해 보세요.

````python
months = ['January', 'February', 'March', 'April', 'May', 'June', 
            'July', 'August', 'September', 'October', 'November', 'December']
avg_temperatures = [-1, 0, 5, 10, 15, 20, 25, 23, 19, 12, 5, 0]
rainfall = [50, 40, 60, 80, 100, 120, 150, 140, 110, 90, 70, 60]
math_scores = [85, 90, 88, 76, 95, 89, 72, 84, 91, 87]
english_scores = [78, 82, 85, 80, 88, 90, 84, 79, 85, 83]
````

````python
plt.plot(months,avg_temperatures)
plt.xticks(rotation= 45)
plt.title("Monthly_avg_Temperatures")
plt.show()
````

````python
plt.bar(months,rainfall)
plt.xticks(rotation= 45)
plt.title("Monthly_Rainfall")
plt.show()
````

````python
data = math_scores , english_scores
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.boxplot(math_scores)
plt.xlabel('Math')
plt.subplot(1,2,2)
plt.boxplot(english_scores)
plt.xlabel('English')
plt.show()
````

````python
plt.hist(math_scores,bins=15,color='blue',edgecolor='black')
plt.show()
````

````python
plt.scatter(math_scores,english_scores)
plt.title('Score')
plt.show()
````

````python
plt.figure(figsize=(20, 15))

plt.subplot(3,3,1)
plt.plot(months,avg_temperatures)
plt.xticks(rotation= 45)
plt.title("Monthly_avg_Temperatures")

plt.subplot(3,3,2)
plt.bar(months,rainfall)
plt.xticks(rotation= 45)
plt.title("Monthly_Rainfall")

plt.subplot(3,3,3)
data = math_scores , english_scores
plt.boxplot(data, tick_labels=['Math', 'English'])

plt.subplot(3,3,4)
plt.hist(math_scores,bins=15,color='blue',edgecolor='black')
plt.subplot(3,3,5)
plt.scatter(math_scores,english_scores)

plt.title('Score')

plt.tight_layout()
plt.show()

````
