```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  # numpy 직접 가져오기

# 예제 데이터 생성
dates = pd.date_range('2026-01-01', periods=100)
prices = pd.Series(range(100)) + np.random.randn(100).cumsum() 
stock_data = pd.DataFrame({'Date': dates, 'Price': prices})

stock_data
```

```python
sns.lineplot(x='Date', y='Price',data = stock_data)
plt.title('Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터셋 로드
tips = sns.load_dataset('tips')
tips.head()
```

```python
# barplot의 기본 동작 : mean() 인데 estimator = sum로 총합 등을 확인

sns.barplot(x='day',y='tip',data=tips,estimator=sum)
plt.title('Total Tips by Day')
plt.xlabel('Day of Week')
plt.ylabel('Total Tip Amount')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터셋 로드
tips = sns.load_dataset('tips')

sns.histplot(tips['total_bill'], kde=True, bins=20)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터셋 로드
tips = sns.load_dataset('tips')
tips
```

```python

# 박스플롯 그리기
sns.boxplot(x='sex', y='tip', data=tips)
plt.title('Tip Amount by Sex')
plt.xlabel('Sex')
plt.ylabel('Tip Amount')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 바이올린 플롯 그리기
sns.violinplot(x='sex', y='tip', data=tips)
plt.title('Tip Amount by Sex')
plt.xlabel('Sex')
plt.ylabel('Tip Amount')
plt.show()
```
