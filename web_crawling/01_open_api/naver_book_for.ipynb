```python
import urllib.parse
import urllib.request
import json
import mysql.connector
import os 
from dotenv import load_dotenv

load_dotenv()




```

```python
# api key 설정

client_id = os.getenv("naver_client_id")
client_secret = os.getenv("naver_secret")
client_secret
```

```python
# 검색어 설정
searchText = urllib.parse.quote("축구")
```

```python
url = "https://openapi.naver.com/v1/search/book.json?query=" + searchText + "&display=1"
# 네이버 api 요청 준비
request = urllib.request.Request(url)
#인증 헤더 추가
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
```

```python
# api 요청 및 결과값 json 반환

response = urllib.request.urlopen(request)  # api 호출
response_body = response.read()            # 응답데이터 (byte)
response_body = json.loads(response_body)

response_body
```

```python
total_count = response_body['total']  # 전체 검색 결과 개수
start_num = 1 
loop_count = total_count // 100 + 1  # 100개씩 끊어서 몇 번 반복할지 계산
book_list = []                # 모든 책 데이터를 누적할 리스트  
```

```python
for i in range(loop_count):
    url = "https://openapi.naver.com/v1/search/book.json?query=" + searchText + "&display=100&start=" + str(start_num)
# 네이버 api 요청 준비

    request = urllib.request.Request(url)

    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)

    response = urllib.request.urlopen(request)  # api 호출
    response_body = response.read()            # 응답데이터 (byte)
    response_body = json.loads(response_body) #json => dict

    book_list += response_body['items']    #기존 리스트에 이어붙임
    start_num += 100
    
    # 최대 1000건만 요청
    if start_num > 1000:
        break

len(book_list)
```

```python
from datetime import datetime
with mysql.connector.connect(
    host = "localhost",
    user = "castleuk",
    password = '1234',
    database = 'bookdb'
) as connection :
    
    with connection.cursor() as cursor:
        
        sql = "insert into naver_book"\
            "(book_title,book_image,author,publisher,isbn,book_description,pub_date)" \
            "values (%s,%s,%s,%s,%s,%s,%s)"
        
        for book_info in book_list:

            pub_str = book_info.get('pubdate','')    # 원본 문자열 가져오기

            if pub_str:        #pubdate 비어있지않으면 data형식으로 변환
                pub_date = datetime.strptime(pub_str, '%Y%m%d').date()
            else:          #비어있으면 None
                pub_date = None

            values = (
                book_info['title'],
                book_info['image'],
                book_info['author'],
                book_info['publisher'],
                book_info['isbn'],
                book_info['description'],
                pub_date
            )

            cursor.execute(sql, values)   #insert문 실행

        # 
        connection.commit()  # 실제 db상 저장
```
