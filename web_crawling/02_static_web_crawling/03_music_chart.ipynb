```python
import requests                          # HTML요청
from bs4 import BeautifulSoup            # HTML 파싱 도구
from urllib.request import urlretrieve   # 이미지 다운로드용
from datetime import datetime            # 파일명 시간 생성용
```

```python
# 스크랩할 음악 정보를 담을

class MusicEntry:
    def __init__(self,title,artist,img_path):
        self.title = title       # 곡 제목
        self.artist = artist         # 자수 명
        self.img_path = img_path # 이미지 경로

    def __repr__(self):
        return f"🎶{self.artist}의 {self.title}🎶 | {self.img_path}"
    
```

```python
#벅스 차트

response = requests.get("https://music.bugs.co.kr/chart")

response
```

```python
# 파싱

bs = BeautifulSoup(response.text, 'html.parser')


bs.prettify()

```

```python
# 곡리스트 정보 (table > tbody > tr)

trach_list = bs.select('table.trackList tbody tr')

result_list = []
trach_list
```

```python
for i, song in enumerate(trach_list[:50]):
    
    title= song.select_one('p.title a').text.strip()
    aritist = song.select_one('p.title a').text.strip()
    img_path = song.select_one('a.thumbnail img')['src']

    # 파일명 : 
   
    file_name = f"../album_images/{datetime.now().strftime("%y%m%d_%H%M%S_")}_{i+1}.jpg'"

    urlretrieve(img_path, file_name)  # 이미지 저장

    # musicentry 객체 생성 후 리스트에 저장
    music_entry = MusicEntry(title,aritist,img_path)
    result_list.append(music_entry)

for result in result_list:
    print(result)
```
