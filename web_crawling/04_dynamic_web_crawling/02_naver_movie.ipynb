```python
# -- 셀레니움 관련 라이브러리

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service


# -- 시간관련 라이브러리
import time
from datetime import datetime            # 파일명 날짜시간 추가

# -- 다운로드 관련 라이브러리
from urllib.request import urlretrieve   # 이미지 다운로드
```

```python
# 크롬 브라우저 실행

service = webdriver.chrome.service.Service("chromedriver.exe")  # 객체생성
driver = webdriver.Chrome(service = service)                    # 크롬 브라우저 실행
```

```python
driver.get("https://naver.com")
time.sleep(1)
```

```python
search_box = driver.find_element(By.ID, 'query')

search_box.send_keys("현재 상영 영화")
search_box.send_keys(Keys.RETURN)
time.sleep(1)
```

```python
for i in range(9):
    movie_items = driver.find_elements(By.CSS_SELECTOR,'div.data_area')
    for movie_item in movie_items:
        # 제목이 들어있는 a 태그의 글자만 추출
        title = movie_item.find_element(By.CSS_SELECTOR,'div.area_text_box > a').text

        #이미지 + 링크 포함된 요소
        img_box = movie_item.find_element(By.CSS_SELECTOR,'a.img_box')
        
        #해당 영화상세페이지 링크
        link = img_box.get_attribute('href')

        # 이미지 소스 경로
        img_src = img_box.find_element(By.TAG_NAME,"img").get_attribute('src')

        print(f"제목: {title}, 상세페이지 : {link}, 이미지 url: {img_src}")

        file_name = f"../movie_images/{datetime.now().strftime("%y%m%d_%H%M%S_")}_{title}.jpg"
        urlretrieve(img_src, file_name)

    if i != 8:
        next_btn = driver.find_element(By.CSS_SELECTOR ,'a._next')
        next_btn.click()
        time.sleep(1)

driver.quit()
```
