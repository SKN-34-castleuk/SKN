# 활성화 함수
Sigmoid : 이진 분류 출력층에서 사용 (0~1)  
Tanh : 순환 신경망(RNN) 등 시계열 모델에서 중심화된 출력값이 필요할때 사용 (-1 ~ 1)    
ReLU : 대부분의 DNN 모델 은닉층에서 기본값으로 사용  
Leaky ReLU : Dead ReLU 문제 보완한 ReLU 계열  

```python
import torch
import torch.nn as nn 
from matplotlib import pyplot as plt
```

```python
torch.sort(torch.randn(100)* 10)
```

```python
# 평균 0, 표준편차 1인 100개의 랜덤 데이터에 10을 곱해 표준편차 10으로 바꾸고, values만 가져온다.
x = torch.sort(torch.randn(100)* 10)[0]   
  
x
```

# Sigmoid

```python
act = nn.Sigmoid()  # 각 입력값을 0~1사이 확률값으로 매핑 (객체 지향 방식)

print(act(x))
```

```python
torch.sigmoid(x)  # 힘수 호출 방식
```

```python
# matplotlib 시각화시 Tensor형태로는 인식이 안되어 numpy형태로 변환을 시켜서 시각화

plt.plot(x.numpy(),torch.sigmoid(x).numpy())
plt.show()
```

# Tanh

```python
act = nn.Tanh()
print(act(x))
```

```python
torch.tanh(x)  # 힘수 호출 방식
```

```python
# matplotlib 시각화시 Tensor형태로는 인식이 안되어 numpy형태로 변환을 시켜서 시각화

plt.plot(x.numpy(),torch.tanh(x).numpy())
plt.show()
```

# ReLU

```python
act = nn.ReLU()
print(act(x))
```

```python
torch.relu(x)  # 힘수 호출 방식
```

```python
# matplotlib 시각화시 Tensor형태로는 인식이 안되어 numpy형태로 변환을 시켜서 시각화

plt.plot(x.numpy(),torch.relu(x).numpy())
plt.show()
```

# Leaky ReLU

```python
act = nn.LeakyReLU(negative_slope= 0.1)
print(act(x))
```

```python
import torch.nn.functional as F

F.leaky_relu(x,negative_slope=0.1)  # 힘수 호출 방식
```

```python
# matplotlib 시각화시 Tensor형태로는 인식이 안되어 numpy형태로 변환을 시켜서 시각화

plt.plot(x.numpy(),F.leaky_relu(x).numpy())
plt.show()
```

```python
plt.plot(x.numpy(),act(x).numpy())
plt.show()
```
