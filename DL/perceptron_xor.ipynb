```python
import torch
import torch.nn as nn
import torch.optim as optim

# XOR 문제 & 답
x = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
y = torch.FloatTensor([[0],[1],[1],[0]])

print(x.size(),y.size())
```

```python

model = nn.Sequential(
    nn.Linear(2,1),     # 입력층 -> 출력층
    nn.Sigmoid()        # 출력값을 0~1 사이의 확률로 변환
)

model
```

```python
# 이진 분류 손실 함수
criterion = nn.BCELoss()
```

```python
# 최적화 함수 정의
optimizer = optim.SGD(
    model.parameters(),     # 학습할 모델 파라미터
    lr = 0.1
)
```

```python
for epoch in range(5000):
    optimizer.zero_grad() # 파라미터 초기화

    output = model(x)               # 예측값 계산
    loss = criterion(output, y)     # 손실 계산

    loss.backward()                 # 기울기 계산
    optimizer.step()                # 가중치 업데이트|

    if epoch % 500 == 0:
        print(loss)
        print(f"Epoch : {epoch}, Loss : {loss.item():.4f}")     # item() : 텐서 내 값을 Python 숫자형태로 꺼냄

```

```python
# 평가 시에는 기울기 계산을 하지 않도록 torch.no.grad() 사용

with torch.no_grad():
    probability = model(x)      # 예측 확률 계산
    prediction = (probability >= 0.5).float()   # 확률이 0.5 이상이면 1, 아니면 0

    print(f"probability : {probability}")
    print(f"prediction : {prediction}")
```

단층 퍼셉트론 모델로는 XOR 문제를 해결하지 못한다.

```python
# 다층 퍼셉트론 모델 정의
model = nn.Sequential(
    nn.Linear(2,4),     # 입력 2개 -> 은닉층 4개
    nn.Tanh(),          # 비선형 활성화 함수

    nn.Linear(4,1),     # 은닉층 4게 -> 출력층 1개
    nn.Sigmoid()        # 출력값을 0~1 사이의 확률로 변환
)

model
```

```python
# 이진 분류 손실 함수
criterion = nn.BCELoss()
```

```python
# 최적화 함수 정의
optimizer = optim.SGD(
    model.parameters(),     # 학습할 모델 파라미터
    lr = 0.1
)
```

```python
for epoch in range(10000):
    optimizer.zero_grad() # 파라미터 초기화

    output = model(x)               # 예측값 계산
    loss = criterion(output, y)     # 손실 계산

    loss.backward()                 # 기울기 계산
    optimizer.step()                # 가중치 업데이트|

    if epoch % 1000 == 0:
        print(f"Epoch : {epoch}, Loss : {loss.item():.4f}")     # item() : 텐서 내 값을 Python 숫자형태로 꺼냄

```

Tanh 활성화 함수 relu계열 함수와의 차이는 초기 학습 안정성이 높고, gradient 흐름에 도움이 되는 장점.
