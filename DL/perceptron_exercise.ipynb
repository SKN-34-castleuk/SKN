```python
import numpy as np

def perceptron(x1, x2, weights, bias):
    x = np.array([x1, x2])  # 입력값을 하나의 vector로 묶음
    w = np.array(weights)   # 가중치 리스트를 백터로 변환
    return int(np.dot(x, w) + bias > 0)
    # return 1 if (w1*x1 + w2*x2 + b) > 0 else 0

# AND Gate
print("AND:", perceptron(1, 1, [0.5, 0.5], -0.7))

# NAND Gate
print("NAND:", perceptron(1, 1, [-0.5, -0.5], 0.7))

# OR Gate
print("OR:", perceptron(1, 0, [0.5, 0.5], -0.2))
```

```python
def weighted_sum(inputs, weights, bias):
    return np.dot(inputs, weights) + bias

inputs = [1, 0]
weights = [0.8, -0.5]
bias = 0.2
output = weighted_sum(inputs, weights, bias)
print("Weighted Sum:", output)
```

```python
from sklearn.neural_network import MLPClassifier

# XOR 문제 해결
X = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [0, 1, 1, 0]

mlp = MLPClassifier(
    hidden_layer_sizes=(5,10), # 은닉층 3개층 뉴런 5게
    activation='relu',  # relu 활성화 함수 (가장 보편적으로 사용하는 활성화 함수)
    max_iter=1000)      # epochs
mlp.fit(X, y)

print("Prediction:", mlp.predict(X))
```

activation : 은님층에서 사용할 활성화 함수
- 입력에 비선형성을 추가해 복잡한 패턴 인식하는 함수

- ReLY 계열
    - relu : 가장 일반적으로 사용, 학습 빠름
    - Leaky ReLu : 음수 영역에 작은 기울기 허용 (죽은 ReLU문제 완화)
    - ELU, GELU : Transformer 계열에서 사용

- tanh : 출력 범위가 -1~1로 니오는게 의미있을 때 사용
    - 입력/출력이 0을 중심으로 대칭
    - RNN계열 (LSTM, GRU)

- logistic : 시그모이드 함수 (초기 신경망)


- 출력층에서 사용할 활성화 함수와 구별을 잘 해야 한다.
이진 분류 : sigmoid (0~1) -> 확률로 해석  
다중 분류 ㅣ softmax (출력범위 0~1. 총합이 1) -> 각 클래스에 대한 확률 분포  
회귀 : linear(없음) -> 예측값 그대로 사용
