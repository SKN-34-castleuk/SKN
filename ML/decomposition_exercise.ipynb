```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# 데이터 로드
iris = load_iris()
X = iris.data

# PCA 적용 (2차원으로 축소)
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

print("원본 데이터 형태:", X.shape)
print("축소된 데이터 형태:", X_reduced.shape)
X
```

```python
X_reduced
```

```python
from sklearn.decomposition import KernelPCA

# 커널 PCA 적용 (RBF 커널 사용)
kpca = KernelPCA(
    n_components=2, # 차원수
    kernel='rbf',   # 가우시안 커널
    gamma=15        # 크면 근처 점에 민감, 작으면 멀리 점에 민감
                 )
X_kpca = kpca.fit_transform(X)

print("커널 PCA로 축소된 데이터 형태:", X_kpca.shape)
X_kpca
```

커널 PCA는 일반 PCA와 달리 역변환 하기 어렵다

```python
from sklearn.manifold import LocallyLinearEmbedding

# LLE 적용
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_lle = lle.fit_transform(X)

print("LLE로 축소된 데이터 형태:", X_lle.shape)
```

## PCA : 클래스의 분산을 가장 크게 보존하는 축
## LDA : 클래스를 가장 잘 구분하는 출 => 분류가 목적일떄 주로 사용


- 비지도학습 + 차원축소 => PCA  
- 지도학습 + 분류 + 차원축소 => LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# 데이터 로드
iris = load_iris()
X = iris.data
y = iris.target

# LDA 적용 (2차원으로 차원 축소)
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)

# LDA 결과 시각화
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis')
plt.title('LDA on Iris Dataset')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.show()
```
